```python
import pandas as pd
import json
from datetime import datetime
```

```python
file_pattern = "data/spotify_chart_202411*.json"

df = []

for date in range(1, 25):
    file = file_pattern.replace("*", str(date).zfill(2))
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        for region, tracks in data.items():
            for track_info in tracks:
                track = track_info["track"]
                artists = ", ".join([artist["name"] for artist in track["artists"]])
                df_row = {
                    "region": region,
                    "track_name": track["name"],
                    "track_id": track["id"],
                    "album_name": track["album"]["name"],
                    "album_id": track["album"]["id"],
                    "artists": artists,
                    # "added_at": track_info["added_at"].split("T")[0]
                    "date": datetime.strptime(f"2024-11-{date}", "%Y-%m-%d")
                }
                df.append(df_row)

combined_df = pd.DataFrame(df)
combined_df
```

```python
combined_df.to_csv("spotify_charts_1124.csv", index=False)
```
